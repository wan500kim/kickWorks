```python
import pandas as pd
from konlpy.tag import Okt
#csv파일 불러오기
df = pd.read_csv('hate_speech_raw_data.csv', encoding = 'cp949')
df.head()
```

```python
#특수문자와 노이즈 제거
okt = Okt()
for i in range(len(df['hate_speech'])):
  sentence = ''
  try:
    for word in okt.morphs(df['hate_speech'][i]):
      sentence = sentence + ' ' + word
      df['hate_speech'] = df['hate_speech'].str.replace(pat=r'[^\w]', repl=r' ', regex=True)
      df['hate_speech'] = df['hate_speech'].str.replace(pat=r'ㅋ', repl=r' ', regex=True)
      df['hate_speech'] = df['hate_speech'].str.replace(pat=r'ㅎ', repl=r' ', regex=True)
      df['hate_speech'] = df['hate_speech'].str.replace(pat=r'ㅡ', repl=r' ', regex=True)
      df['hate_speech'] = df['hate_speech'].str.replace(pat=r'ㅠ', repl=r' ', regex=True)
      df['hate_speech'] = df['hate_speech'].str.replace(pat=r'ㅜ', repl=r' ', regex=True)
      df['hate_speech'] = df['hate_speech'].str.replace(pat=r'ㄷ', repl=r' ', regex=True)
      df['hate_speech'] = df['hate_speech'].str.replace(pat=r'ㅇㄱㄹㅇ', repl=r' ', regex=True)
      df['hate_speech'] = df['hate_speech'].str.replace(pat=r'ㄱ', repl=r' ', regex=True)
  except:
    pass
  df['hate_speech'][i] = sentence
  print(i, df['hate_speech'][i])
  df.to_csv('hate_speech_raw_data.csv', index=False)
```

```python
#각 문장에 한글만 남긴 후 형태소 단위로 분해
okt = Okt()
for i in range(len(df['hate_speech'])):
  sentence = ''
  try:
    for word in okt.nouns(df['hate_speech'][i]):
      sentence = sentence + ' ' + word
  except:
    pass
  df['hate_speech'][i] = sentence
  print(i, df['hate_speech'][i])
  df.to_csv('hate_speech_raw_data.csv', index=False)
```
